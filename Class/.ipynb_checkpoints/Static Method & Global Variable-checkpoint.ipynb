```python
# 3 types of methods. These are instance, class and static

class Student:
    
    school = 'mediu'
    
    def __init__(self, m1, m2, m3):
        self.m1 = m1
        self.m2 = m2
        self.m3 = m3
        
    def avg(self):                             # instance method
        return (self.m1 + self.m2 + self.m3)/3
    
    @classmethod                               # class methods
    def getSchool(school_name):
        return school_name.school
    
    @staticmethod                              # static methods
    def info():
        print("This is from static method. For printing more information")
        

def main():
    s1 = Student(34, 53, 76)
    s2 = Student(85, 23, 66)

    print(s1.avg())
    print(s2.avg())

    print(Student.getSchool())

    Student.info()
    
if __name__=='__main__':
    main()
```

```python
class Donation:
    organization_name = "Motivation Group"                                      # Global Variable
    total_donation = 0                                                          # Global Variable

    def __init__(self, name="No Name", address="No Address", donation=0):
        self.name = name
        self.address = address

        if(donation > 0):
            self.donation = donation
            Donation.total_donation += donation
        else:
            self.donation = 0

    @staticmethod
    def getOranizationName():                                                    # static method
        return Donation.organization_name

    @staticmethod
    def getTotal_donation():                                                     # static method
        return Donation.total_donation

    @staticmethod                                                                # static method
    def setTotal_donation(newTotalDonation):
        Donation.total_donation += newTotalDonation

    def getName(self):
        return self.name

    def setName(self, name):
        self.name = name

    def getAddress(self):
        return self.address

    def setAddress(self, address):
        self.address = address

    def getDonation(self):
        return self.donation

    def setDonation(self, donation):
        if(donation > 0):
            self.donation += donation
            Donation.total_donation += donation
        else:
            self.donation = 0

    def display(self):
        print("Name: " + self.name)
        print("Address: " + self.address)
        print("Donation:", self.donation)



if __name__ == '__main__':
    print("\t\t\t### Organization name: " + Donation.getOranizationName() + " ###")        # static method

    tuhin = Donation('tuhin', 'desa petaling', 600)
    tuhin.display()
    print("\t\t\t###Total Donation:", Donation.getTotal_donation(), "\t###")              # static method

    tuhin.setDonation(250)
    tuhin.setDonation(250)
    tuhin.display()
    print("\t\t\t###Total Donation:", tuhin.getTotal_donation(), "\t###")                 # static method

    nazrul = Donation("nazrul", 'Mantari', 900)
    nazrul.setAddress("Shah Alam")
    nazrul.display()

    Donation.setTotal_donation(1500)                                                      # static method

    print("\t\t\t###Total Donation:", Donation.getTotal_donation(), "\t###")              # static method
```

## Global Variable

```python
class Kettle:
    power_source = "Electricity"         # this is Global variable

    def __init__(self,make, price):
        self.make = make
        self.price = price
        self.on = False

    def switch_on(self):
        self.on = True
        
kenwood = Kettle("Kenwood", 8.99)
hamilton = Kettle("Hamilton", 14.50)

print(Kettle.power_source)
print(kenwood.power_source)
print(hamilton.power_source)
      
print('*' * 80)      

kenwood.power_source = "gas"
print(Kettle.power_source)
print(kenwood.power_source)
print(hamilton.power_source)
      
print('*' * 80)      

hamilton.power_source = "water"
print(Kettle.power_source)
print(kenwood.power_source)
print(hamilton.power_source)

print('*' * 80)

Kettle.power_source = "atomic"
print(Kettle.power_source)
print(kenwood.power_source)
print(hamilton.power_source)

print(Kettle.__dict__)
print(kenwood.__dict__)
print(hamilton.__dict__)
```
