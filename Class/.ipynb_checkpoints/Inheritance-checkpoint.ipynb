## Inheritance - Exercise:1

```python
class A():
    def feature1(self):
        print('Feature 1 working')
    def feature2(self):
        print('Feature 2 working')
        
class B(A):
    def feature3(self):
        print('Feature 3 working')
    def feature4(self):
        print('Feature 4 working')
        
class C(B):
    def feature5(self):
        print('Feature 5 working')
        

if __name__== '__main__':
    a = A()
    b = B()
    c = C()

    a.feature1()
    a.feature2()

    b.feature1()
    b.feature2()
    b.feature3()
    b.feature4()

    c.feature1()
    c.feature2()
    c.feature3()
    c.feature4()
    c.feature5()     
```

```python
class A():
    def __init__(self):
        print('in A Init')                 # This is the constructor
    
    def feature1(self):
        print('Feature 1 working')
    def feature2(self):
        print('Feature 2 working')
        
class B(A):
    def __init__(self):
        A.__init__(self)
        print('in B Init')                 # This is the constructor
    
    def feature3(self):
        print('Feature 3 working')
    def feature4(self):
        print('Feature 4 working')
        
class C(B):
    def __init__(self):
        B.__init__(self)
        print('in C Init') 
        
    def feature5(self):
        print('Feature 5 working')
        

if __name__== '__main__':
    b = B()
    c = C()     
```

```python
class A:
    def feature1(self):
        print("Feature 1 working")
        
        
class B:
    def feature2(self):
        print("Feature 2 working")
        
        
class C(A):                            # single level inheritance
    def feature3(self):
        print("Feature 3 working")
        
        
class D(C):                            # multiple level inheritance
    def feature4(self):
        print("Feature 4 working")
        
        
class E(A, B):                         # multiple inheritance
    def feature5(self):
        print("Feature 5 working")
        

if __name__=='__main__':
    a = A()
    b = B()
    c = C()
    d = D()
    e = E()

    c.feature1()

    d.feature1()
    d.feature3()

    e.feature1()
    e.feature2()
```

## Inheritance - Exercise:2

```python
class Animal():
    def __init__(self):
        print("ANIMAL CREATED")
        
    def who_am_i(self):
        print("I am an Animal")
        
    def eat(self):
        print("I am eating")
        

class Dog(Animal):
    def __init__(self):
        Animal.__init__(self)
        print("Dog Created")

        
if __name__== '__main__':
    myanimal = Animal()
    myanimal.eat()
    myanimal.who_am_i()   

    print()

    myDog = Dog()
    myDog.eat()
    myDog.who_am_i()     
```

## Inheritance - Full

```python
class Animal():
    species = 'mammal'

    def __init__(self, name, weight, speed):
        print('I am from Animal')

        self.name = name
        self.weight = weight
        self.speed = speed

        self.animal_introduction()

    def eat(self):
        return 'I can eat'

    def move(self):
        print('From Animal Move')
        print('My speed: ', self.speed, 'KMPH')

    def animal_introduction(self):
        print('My species name:', Animal.species)
        print('My name is: ', self.name)
        print('My Weight is: ', self.weight)
        print(self.eat())
        self.move()


class Human(Animal):
    def __init__(self, name, weight, has_tail, teeth, speed):
        Animal.__init__(self, name, weight, speed)

        self.has_tail = has_tail
        self.teeth = teeth
        self.speed = speed

        self.human_introduction()

    def talk(self):
        print('I can talk')

    def move(self):
        print('From Human Move')
        print('My speed: ', self.speed, 'KMPH')

    def human_introduction(self):
        print('I have', self.teeth, 'teeth')
        if self.has_tail:
            print('I have Tail')

        self.talk()


class Super_Human(Human):
    def __init__(self, name, weight, has_tail, teeth, speed, can_fly):
        Human.__init__(self, name, weight, has_tail, teeth, speed)
        self.can_fly = can_fly
        self.speed = speed

        self.super_human_introduction()

    def move(self):
        print('From Super Human Move')
        print('My speed: ', self.speed, 'KMPH')

    def super_human_introduction(self):
        if self.can_fly:
            print('I can Fly')
        else:
            print('I am more stronger than human')


if __name__== '__main__':
    superman = Super_Human('James Bond', 82, False, 32, 150, True)
    print(superman.teeth)
    superman.move()

    print('\n')
    human = Human(name='tuhin', weight=64, has_tail=False, teeth=30, speed=15)
    print(human.weight)
    human.move()    
```

### Parameter object passing

```python
class Animal():
    
    def __init__(self, name):
        self.name = name
        
    def speak(self):
        raise NotImplementedError("Error of abstract method")
        
class Dog(Animal):
    
    def speak(self):
        return self.name + ' says woof'
    
class Cat(Animal):
    
    def speak(self):
        return self.name + ' says meow'
    

def pet_speak(pet):
    print(pet.speak())   

    
if __name__== '__main__':
    fido = Dog("Fido")
    isis = Cat('isis')    

    print(fido.speak())
    print(isis.speak())
    
    print()
    
    pet_speak(fido)
    pet_speak(isis)        
```

## Non-Inheritance linking

```python
class Car():
    def __init__(self, brand, model, year):
        self.brand = brand
        self.model = model
        self.year = year
        
        self.odometer_reading = 0
        
    def get_descriptive_name(self):
        long_name = str(self.year) + ' ' + self.brand + ' ' + self.model
        return long_name.title()
    
    def update_odometer(self, mileage):
        if mileage >= self.odometer_reading:
            self.odometer_reading = mileage
        else:
            print("You can't roll back an odometer!")
    
    def read_odometer(self):
        print("This car has " + str(self.odometer_reading) + " miles on it.")
        
    def increment_odometer(self, miles):
        self.odometer_reading += miles
        
        
class ElectricCar(Car):
    def __init__(self, brand, model, year):
        super().__init__(brand, model, year)
        
        self.battery = Battery()
        

class Battery():
    def __init__(self, battery_size = 70):
        self.battery_size = battery_size
        
    def describe_battery(self):
        print("This car has a " + str(self.battery_size) + " KWh battery")
        
    def get_range(self):
        if self.battery_size == 70:
            range = 240
        elif self.battery_size == 85:
            range = 270
            
        message = "This car can go approximately " + str(range)
        message += " miles on a full charge."
        print(message)
    

if __name__== '__main__':
    my_tesla = ElectricCar('tesla', 'model s', 2017)
    print(my_tesla.get_descriptive_name())

    battery = Battery()
    my_tesla.battery.describe_battery()
    my_tesla.battery.get_range()     
```

# Multiple Inheritance

```python
class Car:
    def __init__(self,wheels=4):
        self.wheels = wheels
        # We'll say that all cars, no matter their engine, have four wheels by default.

class Gasoline(Car):
    def __init__(self,engine='Gasoline',tank_cap=20):
        Car.__init__(self)
        self.engine = engine
        self.tank_cap = tank_cap # represents fuel tank capacity in gallons
        self.tank = 0
        
    def refuel(self):
        self.tank = self.tank_cap
        
    
class Electric(Car):
    def __init__(self,engine='Electric',kWh_cap=60):
        Car.__init__(self)
        self.engine = engine
        self.kWh_cap = kWh_cap # represents battery capacity in kilowatt-hours
        self.kWh = 0
    
    def recharge(self):
        self.kWh = self.kWh_cap
        
class Hybrid(Gasoline, Electric):
    def __init__(self,engine='Hybrid',tank_cap=11,kWh_cap=5):
        Gasoline.__init__(self,engine,tank_cap)
        Electric.__init__(self,engine,kWh_cap)
        
        
prius = Hybrid()
print(prius.tank)
print(prius.kWh)

prius.recharge()
print(prius.kWh)
```
